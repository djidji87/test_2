Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
data = pd.read_csv('ratings.csv')
data1 = pd.read_csv('movies.csv')
data = data[ data.rating == 5.0 ]
data2 = data['movieId'].value_counts().reset_index().head(1)
data2.merge(data1, how='left').title.head()
print('Наш фильм {}'.format(data2.merge(data1, how='left').title))
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd
data = pd.read_csv('power.csv')
data_filter = data[ ((data['country']=='Estonia') | (data['country']=='Latvia') | (data['country']=='Lithuania'))\
                   & ((data['category']==4) | (data['category']==12) | (data['category']==21))\
                   & ((data['year']>=2005) & (data['year']<=2010)) & ((data['quantity']>0))]
print('Суммарное потребление стран Прибалтики по указанным фильтрам составляет {}'.format(data_filter.quantity.sum()))
```
